# Predicting emergency department admissions in New South Wales hospitals using advanced machine learning and climate data

### AIMS

The aim of this project is to use machine learning (ML) algorithms to predict patient admissions at the Emergency Department (ED), using patient and climate data. The project will be split into three stages:

1. Acquisition of historical data from a range of hospitals across NSW spanning a 5 year period, and associated meteorological data for a predetermined geographical area within a reasonable radius of those hospitals.
1. Assembling the data for model training and testing. Logistic regression will act as a baseline model, and Extreme Gradient Boosting (XGB) and Deep Neural Networks (DNN) will be used for comparison.  These models return an output of whether an individual is likely to be admitted to hospital, and the associated probability. 
1. Validation of results and preparation for application in clinical use.

### BACKGROUND

The health system is increasingly becoming overcrowded and the pressure on hospitals is higher than ever. The Australian Medical Association of NSW (AMA(NSW)) reported that 1.5 million people have presented to emergency this year alone, the highest amount ever recorded [1]. The president of the AMA (NSW) stated that *“Our hospitals were never built to cope with these numbers and the system has not been designed to cope as effectively as it could''*. A study from 2016 attempted to create an assistance tool called the Sydney Triage and Admission Risk Tool (START) [2]. The tool would predict whether a patient is admitted after presenting to the ED, helping speed up decision making for the ED staff and ultimately increasing flow and reducing waiting times. The study was based on data collected from the Royal Prince Alfred Hospital’s Emergency Department Data collection registry in NSW from the years 2013-2014 for people 16 years and over [2]. The analysis used logistic regression with stepwise variable selection and least absolute shrinkage and selection operator (LASSO), with the final model having an accuracy of around 80% (AUC of 0.82). The study was expanded to include machine learning (ML) models such as decision trees, naive bayes and neural networks in 2018 [3]. However, even with more advanced ML techniques, the study returned similar accuracy to the original logistic regression model. The success of the START tool has yet to be measured as the next phase is to test whether clinical efficiency is improved by using START.

This project proposes to expand on this research by incorporating meteorological data and expanding the size of the dataset to a five year period (rather than the 2013-2014 period of previous studies). There have been several studies showing that extreme weather events increase hospital admission rates for a number of conditions all around the world [4,5,6,7,8], and with the climate changing at an alarming rate, these effects will only become more prevalent. For example, climate factors such as precipitation and air pollution are all known to exacerbate existing illnesses like asthma [9,10], and high temperatures are associated with increased infectious diseases [11]. There are a few tools that are available which assist ED staff in predicting how many patients are likely to turn up to the ED for a given day [12,13], however none of them account for climate in their prediction models.

**Innovation**

Combining meteorological data with admission data in predicting the probability of a person being admitted to hospital has not been performed before in NSW. Previous research has used meteorological data to predict the number of patients (grouped) admitted in a given time period [4,7,8] in time series forecasting. However, with the flexibility and power of machine learning models, we can transform the data to make it suitable for classification modelling and give predictions at the individual level. Since the overwhelming body of research suggests that weather is a key factor in increasing hospitalisation risk, it could possibly lead to better accuracy in predictions than previous models. Meteorological data is also easily obtainable from the Bureau of Meteorology (BOM). XGB and DNN machine learning techniques are highly advanced, powerful and can handle complex data sets without the restrictions of traditional statistical methods.
